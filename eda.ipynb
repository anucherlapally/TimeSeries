```python
import pandas as pd
```

```python
df_schemes = pd.read_csv('data/mutual_fund_data.csv')
print("Scheme Details Data:")
print(df_schemes.columns)
```

```python
df_schemes['AAUM_Quarter'].unique()
```

```python
df_nav = pd.read_parquet('data/mutual_fund_nav_history.parquet')
df_nav = df_nav.reset_index()
df_nav = df_nav.sort_values(by='Date')
print(df_nav.head())
```

```python
df_nav.info()
```

```python
check_data = df_schemes[df_schemes['Scheme_Code'] == 105280]

display(check_data)
```

```python
debt_schemes = df_schemes[df_schemes['Scheme_Category'] == 'Debt Scheme - Liquid Fund']
debt_growth_schemes = debt_schemes[debt_schemes['Scheme_NAV_Name'].str.contains('Growth')]

display(debt_growth_schemes)
```

```python
debt_growth_nav = df_nav[df_nav['Scheme_Code'].isin(debt_growth_schemes['Scheme_Code'].unique())]
valid_navs = debt_growth_nav[debt_growth_nav['NAV'] >= 0]

display(valid_navs['Scheme_Code'].unique())
```

```python

nav_scheme_df = valid_navs.merge(debt_growth_schemes, on='Scheme_Code', how='left', suffixes=('_nav','_data')) \
                    [['Scheme_Code', 'Date', 'NAV_nav', 'Scheme_NAV_Name', 'Launch_Date', 'Closure_Date']]
```

```python
nav_scheme_df.info()
```

```python
display(nav_scheme_df['Date'].unique())
```

```python
# Ensure datetime
nav_scheme_df['Date'] = pd.to_datetime(nav_scheme_df['Date'])

# Unique sorted dates
dates = pd.Series(sorted(nav_scheme_df['Date'].unique()))

# Full continuous calendar range
full_range = pd.date_range(
    start=dates.min(),
    end=dates.max(),
    freq='D'
)

# Missing dates
missing_dates = full_range.difference(dates)

print(f"Total missing dates: {len(missing_dates)}")
print(missing_dates[:20])
```

```python
missing_dates.day_name().value_counts()
```

```python
duplicates = nav_scheme_df.duplicated(subset=['Scheme_Code', 'Date']).sum()

print(duplicates)
```

```python
agg_nav_df = nav_scheme_df.groupby('Date')['NAV_nav'] \
                            .mean() \
                            .reset_index() \

agg_nav_df.columns = ['Date', 'category_avg_nav']

agg_nav_df = agg_nav_df.sort_values('Date')
```

```python
display(agg_nav_df)
```

```python
agg_nav_df['pct_change'] = (
    agg_nav_df['category_avg_nav']
    .pct_change()
)

agg_nav_df['pct_change'].describe()
```

```python
median_df = nav_scheme_df[nav_scheme_df['Date'].isin(thresholded_dates['Date'])]
median_df = median_df[median_df['Date'].dt.day_of_week < 5]

median_df = median_df.groupby('Date')['NAV_nav'].median().reset_index()
median_df.columns = ['Date', 'category_median_nav']

median_df['pct_change'] = median_df['category_median_nav'].pct_change()
```

```python
median_df['pct_change'].describe()
```

```python
scheme_filter_df = nav_scheme_df[nav_scheme_df['Date'].isin(thresholded_dates['Date'])]
scheme_filter_df = scheme_filter_df[scheme_filter_df['Date'].dt.day_of_week < 5]
```

```python
total_dates = scheme_filter_df['Date'].nunique()
```

```python
scheme_coverage = scheme_filter_df.groupby('Scheme_Code')['Date'].nunique().reset_index(name='active_dates')
```

```python
scheme_coverage['Coverage_ratio'] = scheme_coverage['active_dates'] / total_dates
```

```python
stable_schemes = scheme_coverage[scheme_coverage['Coverage_ratio'] >= 0.95]['Scheme_Code']
```

```python
scheme_filter_df = scheme_filter_df[scheme_filter_df['Scheme_Code'].isin(stable_schemes)]
```

```python
scheme_filter_df = scheme_filter_df.groupby('Date')['NAV_nav'].mean().reset_index()
scheme_filter_df.columns = ['Date', 'category_avg_nav']
scheme_filter_df['pct_change'] = scheme_filter_df['category_avg_nav'].pct_change()
```

```python
scheme_filter_df['pct_change'].describe()
```

```python
scheme_filter_df.sort_values(by='pct_change', ascending=False).head(10)
```
