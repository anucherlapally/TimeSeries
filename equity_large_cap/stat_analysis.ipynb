```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
equity_schema_nav = pd.read_parquet('../data/clean/equity_schema_nav.parqet')
```

```python
equity_schema_nav.columns
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    equity_schema_nav['Date'],
    equity_schema_nav['category_avg_nav']
)

plt.title("Large Cap Category Average NAV")
plt.xlabel("Date")
plt.ylabel("Average NAV")

plt.show()
```

```python
equity_schema_nav['diff_1'] = equity_schema_nav['category_avg_nav'].diff()
```

```python
plt.figure(figsize=(12,5))

plt.plot(
    equity_schema_nav['Date'],
    equity_schema_nav['diff_1']
)

plt.title("Differenced Average NAV")
plt.xlabel("Date")
plt.ylabel("Differenced Average NAV")

plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

plot_acf(
    equity_schema_nav['diff_1'].dropna(),
    lags=40
)

plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_pacf

plot_pacf(
    equity_schema_nav['diff_1'].dropna(),
    lags=40
)

plt.show()
```
